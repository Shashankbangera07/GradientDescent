```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples=4, n_features=1,n_informative=1, n_targets=1, noise=80, random_state=42)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X, reg.predict(X), color = 'red')
```

```python
reg.predict(X)
```

```python
# m = -57.99225681
# b = 0
y_pred = ((-57.99 * X)+ 0).reshape(4)
```

```python
y_pred
```

```python
plt.scatter(X,y)
plt.plot(X, reg.predict(X), color = 'red', label = 'OLS')
plt.plot(X, y_pred, color = 'green', label = 'b = 0')
plt.show()
```

```python
m = -57.99
b = 42
loss_slope = -2 * np.sum(y - m*X.ravel() -b)
loss_slope
```

```python
lr = 0.1

step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
```

```python
b
```

```python
b = 0
m = -57.99
lr = 0.01

epochs = 5

for i in range(epochs):
  loss_slope = -2 * np.sum(y - m * X.ravel() - b)
  b = b - (lr * loss_slope)

  y_pred = m* X + b
  plt.plot(X,y_pred)
plt.scatter(X,y)
```

```python
b = 0
m = -57.99
lr = 0.1

epochs = 5

for i in range(epochs):
  loss_slope = -2 * np.sum(y - m * X.flatten() - b)
  b = b - (lr * loss_slope)

  y_pred = m* X + b
  plt.plot(X,y_pred)
plt.scatter(X,y)
```
